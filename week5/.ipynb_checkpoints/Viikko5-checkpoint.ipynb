```python
import pandas as pd

data = pd.read_csv("majava.tsv", sep="\t")
data.dropna(inplace=True)
data = data.loc[data["ParticipationType"] == "Official", ["Grade", "Score"]]

data.head()
```

```python
data2 = data[(data["Score"] >= 30) & (3 <= data["Grade"]) & (data["Grade"] <= 9)].copy()
```

```python
data2.head()
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression().fit(data2[["Grade"]], data2["Score"])
```

```python
regr.coef_, regr.intercept_
```

```python
data2.loc[:, "regr"] = data2["Grade"]*regr.coef_ + regr.intercept_
```

```python
ax = data2.plot(x="Grade", y="Score", kind="scatter")
data2.plot(x="Grade", y="regr", ax=ax, color="r")
```

```python
s = pd.read_csv("tampere_saa.tsv", sep="\t")
s = s.iloc[:, [2, 5]]
s.columns = ["kk", "C"]
s.loc[:, "C"] = pd.to_numeric(s["C"], errors="coerce")
kks = s.groupby("kk")
kkavg = pd.DataFrame(kks["C"].mean())
kkavg.reset_index(inplace=True)
kkavg.plot(x="kk", y="C", style=".")
```

```python
kkregr = LinearRegression().fit(pd.DataFrame(kkavg.index), kkavg["C"])
kkavg.loc[:,"regr"] = kkavg["kk"]*kkregr.coef_ + kkregr.intercept_
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(kkavg["kk"], kkavg["C"])
xs = np.linspace(1, 12, 120)
plt.plot(xs, xs*kkregr.coef_ + kkregr.intercept_, color="red")
```

```python
pd.concat([kkavg["kk"], kkavg["kk"]*kkavg["kk"]], axis=1)
```

```python
kkregr2 = LinearRegression().fit(pd.concat([kkavg["kk"], kkavg["kk"]*kkavg["kk"]], axis=1), kkavg["C"])

print(kkregr2.coef_)

plt.scatter(kkavg["kk"], kkavg["C"])
xs = np.linspace(1, 12, 120)
ys = xs*xs*kkregr2.coef_[1] + xs*kkregr2.coef_[0] + kkregr2.intercept_
plt.plot(xs, ys, color="red")
```

```python
plt.scatter(kkavg["kk"], kkavg["C"])
xs = np.linspace(1, 12, 120)
ys = kkregr2.predict(np.column_stack([xs, xs*xs]))
plt.plot(xs, ys, color="red")
```

```python
from sklearn.preprocessing import PolynomialFeatures

pol2 = PolynomialFeatures(degree=2, include_bias=False)
pol2kk = pol2.fit_transform(kkavg[["kk"]].values)
kkregr2b = LinearRegression().fit(pol2kk, kkavg["C"])
plt.scatter(kkavg["kk"], kkavg["C"])
xs = np.linspace(1, 12, 120)
ys = kkregr2b.predict(pol2.fit_transform(xs[:, np.newaxis]))
plt.plot(xs, ys, color="red")
```

```python
pol15 = PolynomialFeatures(degree=15, include_bias=False)
pol15kk = pol15.fit_transform(kkavg[["kk"]].values)
plt.scatter(kkavg["kk"], kkavg["C"])
kkregr15 = LinearRegression().fit(pol15kk, kkavg["C"])
xs = np.linspace(1, 12, 120)
ys = kkregr15.predict(pol15.fit_transform(xs[:, np.newaxis]))
plt.plot(xs, ys, color="red")
```

```python
trainBools = np.array([kk % 4 != 0 for kk in kkavg["kk"]])
testBools = np.logical_not(trainBools)
trainBools, testBools
```

```python
trainkk = kkavg[["kk", "C"]][trainBools]
testkk = kkavg[["kk", "C"]][testBools]

pol2trainkk = pol2.fit_transform(trainkk[["kk"]].values)
pol2testkk = pol2.fit_transform(testkk[["kk"]].values)
trainregr2 = LinearRegression().fit(pol2trainkk, trainkk["C"])
plt.scatter(trainkk["kk"], trainkk["C"])
plt.scatter(testkk["kk"], testkk["C"], facecolors='none', edgecolors='b')
xs = np.linspace(1, 12, 120)
ys = trainregr2.predict(pol2.fit_transform(xs[:, np.newaxis]))
plt.plot(xs, ys, color="red")
plt.scatter(testkk["kk"], trainregr2.predict(pol2testkk), color="red")
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(testkk["C"], trainregr2.predict(pol2testkk))
```

```python
D = 9
polD = PolynomialFeatures(degree=D, include_bias=False)
polDtrainkk = polD.fit_transform(trainkk[["kk"]].values)
polDtestkk = polD.fit_transform(testkk[["kk"]].values)
trainregrD = LinearRegression().fit(polDtrainkk, trainkk["C"])
print(mean_squared_error(testkk["C"], trainregrD.predict(polDtestkk)))
plt.scatter(trainkk["kk"], trainkk["C"])
plt.scatter(testkk["kk"], testkk["C"], facecolors='none', edgecolors='b')
xs = np.linspace(1, 12, 120)
ys = trainregrD.predict(polD.fit_transform(xs[:, np.newaxis]))
plt.plot(xs, ys, color="red")
plt.scatter(testkk["kk"], trainregrD.predict(polDtestkk), color="red")
plt.ylim(-50,50)
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets, svm

digits = datasets.load_digits()
images_and_labels = list(zip(digits.images, digits.target))
for index, (image, label) in enumerate(images_and_labels[:4]):
    plt.subplot(2, 4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title(f'Training: {label}')
```

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

classifier = svm.SVC(gamma=0.001)

traindata = data[:n_samples // 2]
trainlabels = digits.target[:n_samples // 2]
classifier.fit(traindata, trainlabels)

testdata = data[n_samples // 2:]
testlabels = digits.target[n_samples // 2:]
predicted = classifier.predict(testdata)
predicted
```

```python
images_and_predictions = list(zip(digits.images[n_samples // 2:], predicted))
for index, (image, prediction) in enumerate(images_and_predictions[:4]):
    plt.subplot(2, 4, index + 5)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title(f'Prediction: {prediction}')
```

```python
mistakeBools = testlabels != predicted
sum(mistakeBools), mistakeBools
```

```python
import numpy as np
np.where(mistakeBools)

mistakeIdxs = np.where(mistakeBools)[0]
mistakeIdxs
```

```python
mistakes = [images_and_predictions[mistakeIdxs[i]] for i in range(4)]
for index, (image, prediction) in enumerate(mistakes):
    plt.subplot(2, 4, index + 5)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title(f'Prediction: {prediction}')
```

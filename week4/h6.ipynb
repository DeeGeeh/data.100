```python
import pandas as pd
```

```python
def montlyStockStats(filename):
    df = pd.read_json(filename)

    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)

    df.index = df.index.to_period("M")

    df = df.groupby(df.index)['close'].agg(['min', 'max', 'mean'])
    df.index.name = None

    return df
```

```python
montlyStockStats('amazon2017.json')
```

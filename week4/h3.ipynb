```python
import pandas as pd
```

```python
pd.read_csv('liikenne.csv', delimiter=';')
```

```python
def filterVehicleCounts(filename, hour, low, high): 
    hourStr = hour.astype(str).zfill(2)
    nextHour = hour + 1
    nextHourStr = nextHour.astype(str).zfill(2)
    targetHour = hourStr + '_' + nextHourStr

    df = pd.read_csv(filename, delimiter=';')
    
    filtered_df = df[
        (df[targetHour]) &
        (df[targetHour] >= low) &
        (df[targetHour] <= high)
    ]

    return filtered_df

```

```python
filterVehicleCounts('liikenne.csv')
```

```python
import pandas as pd
```

```python
def sortVehicleCounts(filename, hour):
    
    hourStr = str(hour).zfill(2)
    nextHourStr = str(hour + 1).zfill(2)
    targetHour = f"{hourStr}_{nextHourStr}"

    df = pd.read_csv(filename, delimiter=';')

    sorted_df = df.sort_values(by=[targetHour, 'pvm'], ascending=[True, False], kind='stable')
    
    return sorted_df
```

```python
sortVehicleCounts('liikenne.csv', 0)
```

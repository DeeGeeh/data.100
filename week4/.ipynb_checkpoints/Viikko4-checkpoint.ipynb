Alla yritetty taulukon df lisäys taulukon df3 alapuolelle aiheutti virheen. Virhe johtuu siitä, että
taulukko df3 sisältää useita samannimisiä sarakkeita (kaksi A-saraketta, kaksi B-saraketta, jne.).

Samannimiset sarakkeet estävät concat-operaation, koska Pandas kohdistaa yhdistettävien taulukoiden
sarakkeet (jos lisäys kohdistuu alle) tai rivit (jos lisäys kohdistuu oikealle) niiden otsakkeiden
perusteella. Edellä esimerkiksi df:n sarakkeen A kohdistus on määrittelemätön: pitäisikö se kohdistaa
df3:n ensimmäisen vai toisen A-sarakkeen kanssa?

```python
pd.concat([df3, df])
```

Muunnetaan df3:n sarakeotsakkeet yksikäsitteisiksi, vaikkapa kirjaimin A, B, ..., H.
Tämän jälkeen edellä yritetty concat-operaatio onnistuu.

```python
df3.columns = list("ABCDEFGH")
```

```python
pd.concat([df3, df])
```

Demonstroidaan vielä lisäksi sarakkeiden kohdistuminen hieman selkeämmin: muutetaan
df3:n sarakeotsakkeet niin, että parittomilla sarakkeilla on otsakkeet A, B, C ja D
ja parittomilla W, X, Y ja Z. Tämän jälkeen tehty samanlainen concat-operaatio
ilmentää, kuinka concat kohdistaa df:n sarakkeet A, B, C ja D df3:n vastaavat
otsakkeet omaavien sarakkeiden kanssa.

```python
df3.columns = list("AWBXCYDZ")
```

```python
pd.concat([df3, df])
```
